```python
import os
import openpyxl
import pandas as pd
import shutil
```

```python
########################################################################
# NEED TO MODIFY FOLDERS, FILENAMES BELOW TO REFLECT FOLDER SETUP
########################################################################
fileDir = r"C:\Users\104092\OneDrive - Grundfos\Documents\10-19 Projects\15 SP Integration\Curve PSD"
file_to_split = r"Populated Curve PSD -imp diameters.xlsx"
filePath = os.path.join(fileDir, file_to_split)

wb = openpyxl.load_workbook(filePath)
```

```python
# Read Curve Header Tab into df to split up 
curve_header_ws = wb['Curve Header Data']
df = pd.read_excel(filePath, sheet_name='Curve Header Data', header=8, skiprows=[9,])
```

```python
# Creates list of dataframes split up into groups of "num_tabs_split"
num_tabs_split = 50
list_of_dfs = [df.loc[i:i+num_tabs_split-1,:] for i in range(0, len(df), num_tabs_split)]   
print(f'Total Length of list_of_dfs: {len(list_of_dfs)}')
```

```python
# Approved Tabs List to Keep from mass deleting
approved_tabs = ['Curve Relations Tab', 'Curve Header Data','Curve Header (IEQ use only)']
```

```python
for num, grouped_df in enumerate(list_of_dfs):

    # Create new excel file 
    new_filename = 'SP Group ' + str(num) + '.xlsx'
    workingCopy = os.path.join(fileDir, new_filename)
    shutil.copyfile(filePath, workingCopy)
    new_wb = openpyxl.load_workbook(workingCopy)        

    # Iterate through partnumbers in each sub df,
    # copy corresponding curve tabs to new file
    pn_list = grouped_df['Curve number'].to_list()
    
    for curve_sheet in new_wb.sheetnames:
        if (curve_sheet not in pn_list ) and (curve_sheet not in approved_tabs):
            print(f"DELETE: {curve_sheet}")
            ws_to_delete = new_wb.get_sheet_by_name(curve_sheet)
            new_wb.remove_sheet(ws_to_delete)

    new_wb.save(workingCopy)    
```
